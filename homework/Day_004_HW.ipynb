```python
def MissingData(data):
    #datacount = data.sum()
    total = data.isnull().sum()
    percent = (data.isnull().sum()/data.isnull().count()*100)
    form = pd.concat([total,percent],axis=1,keys=['Total Missing','Missing Data Percent'])
    types=[]
    for col in data.columns:
        dataType=str(data[col].dtype)
        types.append(dataType)
    form['Type'] = types
    return(np.transpose(form))

MissingData(app_train)
```

```python
import matplotlib.pyplot as plt

def PyplotlibScatter(data,tittle):
    plt.scatter(data['CNT_CHILDREN'],data['AMT_REQ_CREDIT_BUREAU_HOUR'],marker='s',s=50)
    
    plt.title(str(tittle))
    plt.legend(loc = 'upper right')
    plt.show()

def PyplotlibPlot(data,tittle):
    plt.plot(data['SK_ID_CURR']/10000,data['AMT_INCOME_TOTAL']/10000)
    plt.title(str(tittle))
    plt.show()
    
PyplotlibScatter(app_train,'Scatter')

PyplotlibPlot(app_train,'Plot')
```
