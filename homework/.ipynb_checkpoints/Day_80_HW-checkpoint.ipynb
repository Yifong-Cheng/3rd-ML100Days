# 請結合前面的知識與程式碼，比較不同的 optimizer 與 learning rate 組合對訓練的結果與影響
### 常見的 optimizer 包含
* SGD
* RMSprop
* AdaGrad
* Adam

```python
import os
import keras

# Disable GPU
os.environ["CUDA_VISIBLE_DEVICES"] = ""
```

```python
train, test = keras.datasets.cifar10.load_data()
```

```python
## 資料前處理
def preproc_x(x, flatten=True):
    x = x / 255.
    if flatten:
        x = x.reshape((len(x), -1))
    return x

def preproc_y(y, num_classes=10):
    if y.shape[-1] == 1:
        y = keras.utils.to_categorical(y, num_classes)
    return y    
```

```python
x_train, y_train = train
x_test, y_test = test

# Preproc the inputs
x_train = preproc_x(x_train)
x_test = preproc_x(x_test)

# Preprc the outputs
y_train = preproc_y(y_train)
y_test = preproc_y(y_test)
```

```python
def build_mlp(input_shape,output_units=10,num_neurons=[512,256,128]):
    """
    Build your own model
    """
    input_layer = keras.layers.Input(input_shape)
    
    for i, n_unit in enumerate(num_neurons):
        if i==0:
            x = keras.layers.Dense(units=n_unit,activation="relu",name="hidden_layer"+str(i+1))(input_layer)
        else:
            x = keras.layers.Dense(units = n_unit,activation="relu",name="hidden_layer"+str(i+1))(x)
    
    out = keras.layers.Dense(units=output_units,activation="softmax",name="output")(x)
    
    model = keras.models.Model(inputs=[input_layer],outputs=[out])
    
    return model
```

```python
## 超參數設定
"""
Set your required experiment parameters
"""
LEARING_RATE=[1e-1,1e-2,1e-3]
EPOCHS=10
BATCH_SIZE=256
MOMENTUM=0.95
#NESTEROV=True

#EPSILON=None
EPSILON=1e-08
DECAY=0.0
#amsgrad=False
BETA_1=0.9
BETA_2=0.999

RHO=0.9
"""
optimizers=[
    keras.optimizers.SGD(lr=LEARING_RATE[0],nesterov=True,momentum=MOMENTUM),
    keras.optimizers.SGD(lr=LEARING_RATE[1],nesterov=True,momentum=MOMENTUM),
    keras.optimizers.SGD(lr=LEARING_RATE[2],nesterov=True,momentum=MOMENTUM),
    keras.optimizers.Adam(lr=LEARING_RATE[0],beta_1=BETA_1, beta_2=BETA_2, epsilon=EPSILON, decay=DECAY, amsgrad=False),
    keras.optimizers.Adam(lr=LEARING_RATE[1],beta_1=BETA_1, beta_2=BETA_2, epsilon=EPSILON, decay=DECAY, amsgrad=False),
    keras.optimizers.Adam(lr=LEARING_RATE[2],beta_1=BETA_1, beta_2=BETA_2, epsilon=EPSILON, decay=DECAY, amsgrad=False),
    keras.optimizers.RMSprop(lr=LEARING_RATE[0], rho=RHO, epsilon=EPSILON, decay=DECAY),
    keras.optimizers.RMSprop(lr=LEARING_RATE[1], rho=RHO, epsilon=EPSILON, decay=DECAY),
    keras.optimizers.RMSprop(lr=LEARING_RATE[2], rho=RHO, epsilon=EPSILON, decay=DECAY)
           ]

optimizers=[
    keras.optimizers.Adam(lr=LEARING_RATE[0],beta_1=BETA_1, beta_2=BETA_2, epsilon=EPSILON, decay=DECAY, amsgrad=False),
    keras.optimizers.Adam(lr=LEARING_RATE[1],beta_1=BETA_1, beta_2=BETA_2, epsilon=EPSILON, decay=DECAY, amsgrad=False),
    keras.optimizers.Adam(lr=LEARING_RATE[2],beta_1=BETA_1, beta_2=BETA_2, epsilon=EPSILON, decay=DECAY, amsgrad=False),
    keras.optimizers.SGD(lr=LEARING_RATE[0],nesterov=True,momentum=MOMENTUM),
    keras.optimizers.SGD(lr=LEARING_RATE[1],nesterov=True,momentum=MOMENTUM),
    keras.optimizers.SGD(lr=LEARING_RATE[2],nesterov=True,momentum=MOMENTUM)    
           ]
#optimizers[0] overfit

"""

optimizers=[
    keras.optimizers.SGD(lr=LEARING_RATE[0],nesterov=True,momentum=MOMENTUM),
    keras.optimizers.SGD(lr=LEARING_RATE[1],nesterov=True,momentum=MOMENTUM),
    keras.optimizers.SGD(lr=LEARING_RATE[2],nesterov=True,momentum=MOMENTUM),
    keras.optimizers.Adam(lr=LEARING_RATE[1],beta_1=BETA_1, beta_2=BETA_2, epsilon=EPSILON, decay=DECAY, amsgrad=False),
    keras.optimizers.Adam(lr=LEARING_RATE[2],beta_1=BETA_1, beta_2=BETA_2, epsilon=EPSILON, decay=DECAY, amsgrad=False),
    keras.optimizers.RMSprop(lr=LEARING_RATE[0], rho=RHO, epsilon=EPSILON, decay=DECAY),
    keras.optimizers.RMSprop(lr=LEARING_RATE[1], rho=RHO, epsilon=EPSILON, decay=DECAY),
    keras.optimizers.RMSprop(lr=LEARING_RATE[2], rho=RHO, epsilon=EPSILON, decay=DECAY)
           ]
optags=["SGD_0","SGD_1","SGD_2","Adam_1","Adam_2","RMSprop_0","RMSprop_1","RMSprop_2"]
```

```python
for op,opt in zip(optimizers,optags):
    print("experment with optag=%s, optimizers %s" %(str(opt),str(op)))
```

```python
results = {}
"""
建立你的訓練與實驗迴圈並蒐集資料
"""
for op,opt in zip(optimizers,optags):
    keras.backend.clear_session()
    print("experment with optimizers %s" %str(opt))
    
    model = build_mlp(input_shape=x_train.shape[1:])
    model.summary()
    
    model.compile(loss="categorical_crossentropy",metrics=["accuracy"],optimizer=op)
    model.fit(x_train, y_train, 
              epochs=EPOCHS, 
              batch_size=BATCH_SIZE, 
              validation_data=(x_test, y_test), 
              shuffle=True)
    train_loss = model.history.history["loss"]
    valid_loss = model.history.history["val_loss"]
    train_acc = model.history.history["accuracy"]
    valid_acc = model.history.history["val_accuracy"]
    
    exp_name_tag = "exp-optimizers-%s " % str(opt)
    results[exp_name_tag] = {'train-loss': train_loss,
                             'valid-loss': valid_loss,
                             'train-acc': train_acc,
                             'valid-acc': valid_acc}
    
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

"""
將實驗結果繪出
"""
color_bar=["red","green","blue","yellow","grey","orange","black","pink","purple","brown"]

plt.figure(figsize=(8,6))
for i, cond in enumerate(results.keys()):
    plt.plot(range(len(results[cond]['train-loss'])),results[cond]['train-loss'], '-', label=cond, color=color_bar[i])
    plt.plot(range(len(results[cond]['valid-loss'])),results[cond]['valid-loss'], '--', label=cond, color=color_bar[i])
plt.title("Loss")
plt.legend()
plt.show()

plt.figure(figsize=(8,6))
for i, cond in enumerate(results.keys()):
    plt.plot(range(len(results[cond]['train-acc'])),results[cond]['train-acc'], '-', label=cond, color=color_bar[i])
    plt.plot(range(len(results[cond]['valid-acc'])),results[cond]['valid-acc'], '--', label=cond, color=color_bar[i])
plt.title("Accuracy")
plt.legend()
plt.show()
```
